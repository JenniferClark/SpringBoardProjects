```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
# Step1: import libaries
import pandas as pd
import requests
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Step2: Fetching Data
base_url = 'https://archive-api.open-meteo.com/v1/archive'
latitude_london = 51.5074
longitude_london = -0.1278
start_date = "2022-01-01"
end_date = "2022-12-31"
hourly_variables = ['precipitation']

# London Data
params_london = {
    'latitude': latitude_london,
    'longitude': longitude_london,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': hourly_variables,
    'timezone': 'Europe/London'
}

response_london = requests.get(base_url, params_london)

if response_london.status_code == 200:
    print('London data fetched successfully!')
else:
    print(f'Failed to fetch London data. Status code: {response_london.status_code}')

data_london = response_london.json()
data_london.keys()

hourly_data_london = data_london['hourly']
df_london = pd.DataFrame(hourly_data_london)
df_london.head()
```

```python
# Seattle Data

latitude_seattle = 47.6062
longitude_seattle = -122.3321

params_seattle = {
    'latitude': latitude_seattle,
    'longitude': longitude_seattle,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': hourly_variables,
    'timezone': 'Europe/London'
}

response_seattle = requests.get(base_url, params_seattle)

if response_seattle.status_code == 200:
    print('Seattle data fetched successfully!')
else:
    print(f'Failed to fetch Seattle data. Status code: {response_seattle.status_code}')

data_seattle = response_seattle.json()

hourly_data_seattle = data_seattle['hourly']
df_seattle = pd.DataFrame(hourly_data_seattle)
df_seattle.head()
```

```python
# Step3: Data Cleaning
# Convert 'time' to datetime format (for easy handling), set it as the index, and handle missing values

# London
df_london['time'] = pd.to_datetime(df_london['time'])
# df_london.head(5)
df_london.set_index('time', inplace=True)
df_london.head(5)

print('Missing values in London data:')
print(df_london.isnull().sum())

df_london.fillna(method='ffill', inplace=True)
df_london.describe()
```

```python
# Seattle
df_seattle['time'] = pd.to_datetime(df_seattle['time'])
# df_london.head(5)
df_seattle.set_index('time', inplace=True)
df_seattle.head(5)

print('Missing values in Seattle data:')
print(df_seattle.isnull().sum())

df_seattle.fillna(method='ffill', inplace=True)
df_seattle.describe()
```

```python
# Step4: resample data

weekly_london = df_london.resample('W').sum()
weekly_seattle = df_seattle.resample('W').sum()

print('Weekly London preview:')
print(weekly_london.head())
print('Weekly Seattle preview:')
print(weekly_seattle.head())
```

```python
# Step5: Visualization 
combined_weekly = pd.DataFrame({
    'London': weekly_london['precipitation'],
    'Seattle': weekly_seattle['precipitation']
})

combined_weekly.plot(figsize=(14, 6), title='Weekly Total Precipitation: London vs Seattle (2022)')
plt.xlabel('Week')
plt.ylabel('Total Precipitation (mm)')
plt.grid(True)
plt.show()
```

```python
# weekly difference
combined_weekly['Difference'] = combined_weekly['London'] - combined_weekly['Seattle']
combined_weekly.head(5)

plt.figure(figsize=(14, 6)) 
plt.plot(combined_weekly.index, combined_weekly['Difference'], label='Seattle - London Precipitation Difference')
plt.title('Weekly Precipitation Difference: Seattle vs London (2022)')
plt.xlabel('Week')
plt.ylabel('Difference (mm)')
plt.axhline(0, color='gray', linestyle='--')  # Horizontal line at 0 for reference (where difference is zero)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Step6: Analysis
total_london = df_london['precipitation'].sum()
total_seattle = df_seattle['precipitation'].sum()

print(f'Total rainfall in London (2022): {total_london: .2f} mm')
print(f'Total rainfall in Seattle (2022): {total_seattle: .2f} mm')

if total_seattle > total_london:
    print('Seattle had more rainfall overall.')
elif total_london > total_seattle:
    print('London had more rainfall overall.')
else:
    print('Both cities had the same total rainfall.')
```

```python
# Interesting patterns (manual observation + simple stats)
print('\nInteresting Patterns:')
print(f"London's max weekly rainfall: {weekly_london['precipitation'].max():.2f} mm")
print(f"Seattle's max weekly rainfall: {weekly_seattle['precipitation'].max():.2f} mm")
print(f"London's average weekly rainfall: {weekly_london['precipitation'].mean():.2f} mm")
print(f"Seattle's average weekly rainfall: {weekly_seattle['precipitation'].mean():.2f} mm")
```

```python
correlation = combined_weekly['London'].corr(combined_weekly['Seattle'])
print(f'Correlation between weekly rainfall in London and Seattle: {correlation:.2f}')
```

```python
# Step7: Saving data

# Save London data
df_london.to_csv('london_weather.csv')

# Save Seattle data
df_seattle.to_csv('seattle_weather.csv')
```
